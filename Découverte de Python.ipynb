# Découverte de Python

Python est un langage de programmation gratuit, évolutif et fortement utilisé par les développeurs d'application et les scientifique.

L'objectif est de vous donner quelques clés pour vous permettre de comprendre et de créer des programmes simples.

## Comme une simple calculatrice

Python peut être utilisé comme une simple calculatrice. On peut alors effectué des calculs sur des nombres entiers, des nombres flottants mais aussi sur des chaines de caractères, des listes de nombres...

**Par exemple :**

```
2 + 5 # une somme entre deux entiers
```



*   *Pour executer la ligne de code précédente, passer la souris sur [ ] et cliquer sur la flêche qui apparaît*
*   *vous aurez constaté que le calcul s'effectue bien sans prendre en compte le commentaire qui suit le symbole #*



**Autres exemples :**

```
2.3 + 5.9 # une somme de deux nombres flottants (à virgules)
```

*un nombre à virgule s'écrit avec un point contrairement à nos habitudes en math.*

```
5 + 6.1 # une somme entre un entier et un flottant
```

```
91 - 201 # la soustraction
```

```
8 / 3 # la division simple
```

```
8 / 4 # une autre division simple
```

*Vous consterez que le résultat de la simple division est toujours un nombre flottant*

```
9 * 123 # la multiplication
```

```
9.23 * 265.32
```

```
23**6 # la puissance
```

```
32665894**256547
```

Comme vous constatez, ça fait beaucoup de chiffres. Le résultat obtenu est la valeur exacte. Cela ne fonctionne pas avec des nombres flottants. Ceux-ci sont affichés avec 15 chiffres significatifs.

```
23.651 * 523.125256
```

**Deux nouveaux opérateurs :**

```
92 // 3 # quotient de la division euclidienne
```

```
92 % 3 # reste de la division euclidienne
```

> 92 = 3 * **30** + **2**

## Notion d'objets et de variables

En Python tout est objet. Un objet fait partie d'une famille d'objets qu'on appelle une **classe**. Ainsi, les nombres entiers sont une classe. Le nombre entier 3 est un objet de cette classe.
Lorsqu'on crée (ou utilise un objet), on définit en même temps son type (sa classe) et toutes les opérations et propriétés qui lui sont associées.


Pour faire un programme, on aura besoin de manipuler des objets : les créer, les faire évoluer au cours du programme, les afficher.

Une **variable** est une référence à un objet. Une variable peut référencer des objets différents au cours de l'éxécution d'un programme.

Pour créer une variable, on utilise une syntaxe **d'affectation** (d'assignation).

```
a = 6  # la variable s'appelle a, elle référence l'objet 6. La relation d'affectation est réalisé avec " = "
```

```
a
```

Pour récupérer le contenu de `a`, il suffit de rappeler la variable comme ci-dessus.

```
a = 13 # on a assigné un nouvel objet à la variable a.
```

```
b = a + 5 # on assigne à la variable b, le résultat du calcul a + 5
```

```
a = a + 3 # on assigne à a son ancienne valeur augmentée de 3.
```

```
"a=", a, "b=", b
```

*Cette dernière opération s'appelle une **incrémentation**.*

Une variable peut aussi référencé un objet texte.

```
c = "un texte" # ici défini avec les doubles-quotes
c
```

```
d = 'un autre texte' # défini avec des simples quotes
d
```

```
e = "deux méthodes, c'est intéressant" # l'apostrophe fait partie du texte, on utilise le guillemet
e
```

## Nom des variables

Autant que possible, éviter des noms de variable d'une seule lettre. Si vous souhaitez écrire un programme clair et compréhensible, donner un vrai nom à votre variable.

```
p = "Christophe" # C'est possible
```

```
prenom = "Christophe" # C'est bien mieux
```

Vous pouvez utiliser tous les mots à partir du moment : 

*   ne commence pas par un chiffre, ni par "_"
*   ne contient pas de symbole autre que " _ "
*   pas d'espace
*   éviter les accents (c'est possible néanmoins)



```
mon_prenom = "Christophe" # une variable bien nommée
```

```
Mon_Prenom = "Michel" # une autre, mais différente
```

```
mon_prenom == Mon_Prenom
```

*Le nom des variables est sensible à la casse.*
